```python
"""
254. Factor Combinations
Numbers can be regarded as product of its factors. For example,
8 = 2 x 2 x 2;
= 2 x 4.
Write a function that takes an integer n and return all possible combinations of its factors.
Note:

	1. You may assume that n is always positive.
	2. Factors should be greater than 1 and less than n.

Examples:
input: 1
output:
[]

input: 37
output:
[]

input: 12
output:
[
[2, 6],
[2, 2, 3],
[3, 4]
]

input: 32
output:
[
[2, 16],
[2, 2, 8],
[2, 2, 2, 4],
[2, 2, 2, 2, 2],
[2, 4, 4],
[4, 8]
]

"""
def getFactors(n):
    todo = [(n, 2, [])]
    combis =[]
    while (todo):
        n,i,combi = todo.pop()
        while i*i <= n:
            if n%i == 0:
                todo += (n/i, i, combi+[i]),
                combis += combi + [i, n/i],
            i += 1
    return (combis)

print ("Iterative Implementation")
print ("Factor for 16 ",getFactors(16))
print ("Factor for 37 ",getFactors(37))
print ("Factor for 5 ",getFactors(5))
print ("Factor for 12 ",getFactors(12))
print ("Factor for 32 ",getFactors(32))
        
     
print ("Recursive Implementation")
def rec_factor (n, i, combi, combis):
    while i*i <= n:
        if n % i == 0:
            combis += combi + [i, n/i],
            rec_factor (n/i, i, combi+[i], combis)
        i += 1
    return combis

def rec_factors (n):
    return rec_factor(n, 2, [], [])
print ("Factor for 16 ",rec_factors(16))
print ("Factor for 37 ",rec_factors(37))
print ("Factor for 5 ",rec_factors(5))
print ("Factor for 12 ",rec_factors(12))
print ("Factor for 32 ",rec_factors(32))
        
```
